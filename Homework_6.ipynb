```python
#PROBLEM 2a
#Use CIFAR-10 dataset
train_data = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transforms.ToTensor())

#Determine mean and standard deviation
img = torch.stack([img_t for img_t, _ in train_data], dim=3)
mean = img.view(3,-1).mean(dim=1)
std_dev = img.view(3,-1).std(dim=1)

#Try to use GPU to speed up training
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

#Normalization transformation definition
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean,std_dev)
])

#Normalize dataset
train_data = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
train_loader = torch.utils.data.DataLoader(train_data, batch_size = 64, shuffle=True) #Batch size recommended
test_data = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
test_loader = torch.utils.data.DataLoader(test_data, batch_size=64, shuffle=False) #Batch size recommended
```

```python
#Model definition, 1 hidden layer size 512
model = nn.Sequential(
    nn.Flatten(),
    nn.Linear(3072, 512), #32*32*3=3072
    nn.Tanh(),
    nn.Linear(512,10) #10 classes
).to(device) #GPU

#Loss/optimizer definition
loss_fn = nn.CrossEntropyLoss() #Recommended by friend, seems to give better results than NLLLoss
optimizer = optim.SGD(model.parameters(), lr=0.001) #SGD usually works best, lr can be changed if needed
```

```python
#Training loop
epochs = 300 #300 required for 2b, so 300 for 2a seems reasonable, change if necessary
sum_start_time = time.time() #For training duration
for epoch in range(epochs):
  start_time = time.time()
  loss_over_time = 0.0
  for inputs, labels, in train_loader:
    inputs, labels = inputs.to(device), labels.to(device)

    optimizer.zero_grad()
    pred = model(inputs)
    loss = loss_fn(pred, labels)
    loss.backward()
    optimizer.step()
    loss_over_time += loss.item()


  #Test
  model.eval()
  pred_correct = 0
  pred_total = 0
  with torch.no_grad(): #No learning form test
    for inputs, labels in test_loader:
      inputs, labels = inputs.to(device), labels.to(device)
      pred = model(inputs)
      _, predicted = torch.max(pred.data, 1)
      pred_total += labels.size(0)
      pred_correct += (predicted == labels).sum().item()
  pred_accuracy = pred_correct/pred_total
  model.train()

  #Training duration
  final_time = time.time()
  train_time = final_time - start_time

  #Display info:
  #Epoch/loss
  if epoch % 10 == 0:
    print(f'Epoch: {epoch+1}/{epochs}, Loss: {loss_over_time/len(train_loader)}, Training Time: {train_time:.2f} seconds')
  #Time
  sum_end_time = time.time()
  sum_train_time = sum_end_time - sum_start_time
  print(f'Total Training Time: {sum_train_time:.2f} seconds')
  #Accuracy
  print(f'Test Accuracy: {pred_accuracy * 100:.2f}%')

#Final Accuracy after training
print(f'Final Test Accuracy: {pred_accuracy * 100:.2f}%')
```

```python
#PROBLEM 2b
#Use CIFAR-10 dataset
train_data = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transforms.ToTensor())

#Determine mean and standard deviation
img = torch.stack([img_t for img_t, _ in train_data], dim=3)
mean = img.view(3,-1).mean(dim=1)
std_dev = img.view(3,-1).std(dim=1)

#Try to use GPU to speed up training
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

#Normalization transformation definition
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean,std_dev)
])

#Normalize dataset
train_data = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
train_loader = torch.utils.data.DataLoader(train_data, batch_size = 64, shuffle=True) #Batch size recommended
test_data = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
test_loader = torch.utils.data.DataLoader(test_data, batch_size=64, shuffle=False) #Batch size recommended
```

```python
#Model definition, 1 hidden layer size 512
model = nn.Sequential(
    nn.Flatten(),
    nn.Linear(3072, 512), #32*32*3=3072
    nn.ReLU(),  #Better results than tanh
    nn.Linear(512,128),
    nn.ReLU(),
    nn.Linear(128,64),
    nn.ReLU(),
    nn.Linear(64,10) #10 classes
).to(device) #GPU

#Loss/optimizer definition
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.001) #SGD usually works best, lr can be changed if needed
```

```python
#Training loop
epochs = 300 #300 required for 2b
sum_start_time = time.time() #For training duration
for epoch in range(epochs):
  start_time = time.time()
  loss_over_time = 0.0
  for inputs, labels, in train_loader:
    inputs, labels = inputs.to(device), labels.to(device)

    optimizer.zero_grad()
    pred = model(inputs)
    loss = loss_fn(pred, labels)
    loss.backward()
    optimizer.step()
    loss_over_time += loss.item()


  #Test
  model.eval()
  pred_correct = 0
  pred_total = 0
  #For confusion matrix
  cf_predicted = []
  cf_labels = []
  with torch.no_grad(): #No learning form test
    for inputs, labels in test_loader:
      inputs, labels = inputs.to(device), labels.to(device)
      pred = model(inputs)
      _, predicted = torch.max(pred.data, 1)
      pred_total += labels.size(0)
      pred_correct += (predicted == labels).sum().item()
      cf_predicted.extend(predicted.cpu().numpy()) #Help says this will allow for confusion matrix
      cf_labels.extend(labels.cpu().numpy())  #Help says this will allow for confusion matrix
  pred_accuracy = pred_correct/pred_total
  model.train()

  #Training duration
  final_time = time.time()
  train_time = final_time - start_time

  #Display info:
  #Epoch/loss
  if epoch % 10 == 0:
    print(f'Epoch: {epoch+1}/{epochs}, Loss: {loss_over_time/len(train_loader)}, Training Time: {train_time:.2f} seconds')
  #Time
  sum_end_time = time.time()
  sum_train_time = sum_end_time - sum_start_time
  print(f'Total Training Time: {sum_train_time:.2f} seconds')
  #Accuracy
  print(f'Test Accuracy: {pred_accuracy * 100:.2f}%')

#Final Accuracy after training
print(f'Final Test Accuracy: {pred_accuracy * 100:.2f}%')

```

```python
#Confusion matrix
cm = confusion_matrix(cf_labels,cf_predicted)
print('Trained Model - Confusion Matrix')
print(cm)
```
